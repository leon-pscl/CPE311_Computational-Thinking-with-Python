## Transform

### handle missing values

```python
commute.isna()
```

```python
commute.count()
```

### Handle duplicates

```python
commute.duplicated().sum
```

### change column names

```python
commute = commute.rename (columns = {
                            'Kepuasan Hidup': 'Life_Satisfaction',
                            'B.1': 'Housing_Affordability',
                            'No. Responden': 'Respondent_ID'
})
#That's all we need to change
```

### Fix data types, if possible

```python
commute.dtypes
```

```python
commute = commute.astype({
    'A.1': 'float',
    'Life_Satisfaction': 'string'
})

commute.dtypes
```
